```python
import pymongo
from pymongo import MongoClient
import json
import tweepy
import twitter
from pprint import pprint
import configparser
import pandas as pd
```

```python
config = configparser.ConfigParser()
config.read('config.ini')
```

```python
CONSUMER_KEY      = config['mytwitter']['api_key']
CONSUMER_SECRET   = config['mytwitter']['api_secrete']
OAUTH_TOKEN       = config['mytwitter']['access_token']
OATH_TOKEN_SECRET = config['mytwitter']['access_secrete']

mongod_connect = config['mymongo']['connection']
```

```python
client = MongoClient(mongod_connect)
```

```python
db = client.test # use or create a database named demo
PS5 = db.PS5 #use or create a collection named tweet_collection
PS5.create_index([("id", pymongo.ASCENDING)],unique = True) # make sure the collected tweets are unique
```

```python
rest_auth = twitter.oauth.OAuth(OAUTH_TOKEN,OATH_TOKEN_SECRET,CONSUMER_KEY,CONSUMER_SECRET)
rest_api = twitter.Twitter(auth=rest_auth)
```
